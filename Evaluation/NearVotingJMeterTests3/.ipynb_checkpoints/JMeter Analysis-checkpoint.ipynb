```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
from IPython.display import Image
```

```python
def load_dataset(results):
    df = pd.read_csv(results, usecols=['timeStamp','elapsed','success','bytes','Latency', 'IdleTime','Connect'])
    df['totalElapsed'] = df.elapsed.cumsum()
    df['throughput'] = ((df.index+1)/(df.totalElapsed/(df.index+1))*60000)
    return df
```

```python
def load_summary(summary):
    df = pd.read_csv(summary)
    return pd.read_csv(summary)
```

```python
add1000 = load_dataset('Add/Add_1000T.csv')
addSum1000 = load_summary('Add/Add_1000T_Sum.csv')
```

```python
vote1000 = load_dataset('Vote/Vote_1000T.csv')
voteSum1000 = load_summary('Vote/Vote_1000T_Sum.csv')
```

```python
get1000b = load_dataset('Get/Get_1000T_50.csv')
getSum1000b = load_summary('Get/Get_1000T_50_Sum.csv')

get1000_250c = load_dataset('Get/Get_1000T_250.csv')
getSum1000_250c = load_summary('Get/Get_1000T_250_Sum.csv')
```

```python
add_compressed = load_dataset('Add/Add_1000T_Compressed.csv')
addSum_compressed = load_summary('Add/Add_1000T_Compressed_Sum.csv')

get_50_compressed = load_dataset('Get/Get_1000T_50_Decompressed.csv')
getSum_50_compressed = load_summary('Get/Get_1000T_50_Decompressed_Sum.csv')

get_250_compressed = load_dataset('Get/Get_1000T_250_Decompressed.csv')
getSum_250_compressed = load_summary('Get/Get_1000T_250_Decompressed_Sum.csv')
```

### Elapsed Time per Transcation across 500 and 1000 Threads

In this section we plot the elapsed time of the add, vote, and get transactions across 500 and 1000 threads respectively. We can see that in the 500 thread plot, there is not a dramatic increase in elapsed time because the backend server does not hang. When we look at the 1000 thread plot we can see a near vertical line in the elapsed time, showing the number of threads that cause a hang in the backend server. 

```python
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(12)
ax.plot(range(0,1000), add1000.elapsed, linewidth=2.0, label="Add")
ax.plot(range(0,1000), vote1000.elapsed, linewidth=2.0, label="Vote")
ax.plot(range(0,1000), get1000b.elapsed, linewidth=2.0, label="Get50c")
ax.plot(range(0,1000), get1000_250c.elapsed, linewidth=2.0, label="Get250c")
ax.title.set_text('Elapsed Time Per Transaction v 1000 Threads')
ax.legend(loc="upper left")
```

```python
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(12)
ax.plot(range(0,1000), add_compressed.elapsed, linewidth=2.0, label="Add")
ax.plot(range(0,1000), vote1000.elapsed, linewidth=2.0, label="Vote")
ax.plot(range(0,1000), get_50_compressed.elapsed, linewidth=2.0, label="Get 50c")
ax.plot(range(0,1000), get_250_compressed.elapsed, linewidth=2.0, label="Get 250c")
ax.title.set_text('Elapsed Time Per Transaction v 500 Threads')
ax.legend(loc="upper left")
```

### Throughput Per Minute over 1000 Threads
In the following section we calculate the average throughput per minute of each transaction across each the add, vote, and get function. We test the get function while retrieving 50 candidates as opposed to 250 candidates. Throughput is calculated as (number of requests)/(total elapsed time). For the following sequence we test the results of compression in the backend function. 

#### LZ-UTF8
In order to incorporate compression, a package called lzutf8 was used on the backend. The data was either compressed before entry into the blockchain or decompressed after retrieval from blockchain. The lzutf8 package incorporates lz77 compression on top of utf8 encoding. The package allows for string compression into into a binary string sequence which allows for transfer of compressed string across json formats. 

One other technique considered was Run length encoding, however due to our data format of storage an object id RLE was not a good fit instead expanding the string stored instead of compressing.  

```python
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(12)
ax.plot(range(0,1000), add1000.throughput, linewidth=2.0, label="Add")
ax.plot(range(0,1000), vote1000.throughput, linewidth=2.0, label="Vote")
ax.plot(range(0,1000), get1000b.throughput, linewidth=2.0, label="Get 50c")
ax.plot(range(0,1000), get1000_250c.throughput, linewidth=2.0, label="Get 250c")
ax.title.set_text('Throughput/Min v Threads without compression')
ax.legend(loc="upper left")
```

```python
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(12)
ax.plot(range(0,1000), add_compressed.throughput, linewidth=2.0, label="Add")
ax.plot(range(0,1000), vote1000.throughput, linewidth=2.0, label="Vote")
ax.plot(range(0,1000), get_50_compressed.throughput, linewidth=2.0, label="Get 50c")
ax.plot(range(0,1000), get_250_compressed.throughput, linewidth=2.0, label="Get 250c")
ax.title.set_text('Throughput/Min v Threads With Compression')
ax.legend(loc="upper left")
```

```python
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(12)
ax.plot(range(0,1000), add_compressed.throughput, linewidth=2.0, label="compressed")
ax.plot(range(0,1000), add1000.throughput, linewidth=2.0, label="normal")
ax.title.set_text('Throughput comparison of \n addCandidate compressed vs non compressed')
ax.legend(loc="upper left")
```

```python
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(12)
ax.plot(range(0,1000), get_50_compressed.throughput, linewidth=2.0, label="compressed")
ax.plot(range(0,1000), get1000b.throughput, linewidth=2.0, label="normal")
ax.title.set_text('Throughput comparison of \n getCandidates 50c compressed vs non compressed')
ax.legend(loc="upper left")
```

```python
fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(12)
ax.plot(range(0,1000), get_250_compressed.throughput, linewidth=2.0, label="compressed")
ax.plot(range(0,1000), get1000_250c.throughput, linewidth=2.0, label="normal")
ax.title.set_text('Throughput comparison of \n getCandidates 250c compressed vs non compressed')
ax.legend(loc="upper left")
```
