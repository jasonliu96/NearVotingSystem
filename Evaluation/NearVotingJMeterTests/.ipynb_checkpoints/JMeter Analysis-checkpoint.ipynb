```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
def load_dataset(results):
    df = pd.read_csv(results, usecols=['timeStamp','elapsed','success','Latency', 'IdleTime','Connect'])
    df['totalElapsed'] = df.elapsed.cumsum()
    df['throughput'] = ((df.index+1)/(df.totalElapsed/(df.index+1))*60000)
    return df
```

```python
def load_summary(summary):
    return pd.read_csv(summary)
```

```python
addCand100 = load_dataset('100tAddCandBase.csv')
addSummary100 = load_summary('100tAddCandBaseSummary.csv')
```
